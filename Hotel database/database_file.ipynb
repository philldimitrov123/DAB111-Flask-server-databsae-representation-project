```python
import sqlite3
import pandas as pd
from pathlib import Path
```

```python
base_path = Path.cwd()
df = pd.read_csv(base_path/'hotel_bookings_cleaned')
#load data


con = sqlite3.connect('hotelDB.db')
#connect/create to database


df.to_sql('hotel_table', con , if_exists='replace', index=False)
#create table


cursor = con.cursor()
#create cursor


query_db = cursor.execute("SELECT * FROM hotel_table LIMIT 25")
get_all = query_db.fetchall()
print(get_all)
#query table

con.commit()
#commit changes
con.close()
#close connection





```
